```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def check(fit, x, y):
    y_fit = fit[2] + fit[1] * x + fit[0] * x ** 2
    plt.figure(dpi=100)
    plt.plot(x, y, '.', markersize=5)
    plt.plot(x, y_fit, lw=1, c='red')
    plt.xlabel('p')
    plt.ylabel('M / N*mm')
    plt.show()
    loss = ((y_fit - y) ** 2).mean()
    print('loss =', loss)
    
    
def fit_curve(path, check_result=True):
    data = pd.read_csv(path)
    x = np.array(data.iloc[:,0]) / 100
    y = np.array(data.iloc[:,1])
    fit = np.polyfit(x, y, 2)
    if check_result:
        check(fit, x, y)
    return fit
```

```python
fit1 = fit_curve('./react1.csv')
print(fit1)
```

```python
fit2 = fit_curve('./react2.csv')
print(fit2)
```

```python
fit3 = fit_curve('./react3.csv')
print(fit3)
```

```python
plt.figure(dpi=100)
x = np.linspace(0,1,30)
y1 = fit1[0]*x**2+fit1[1]*x+fit1[2]
y2 = fit2[0]*x**2+fit2[1]*x+fit2[2]
y3 = fit3[0]*x**2+fit3[1]*x+fit3[2]
plt.plot(x, y1, x, y2, x, y3)
plt.show()
```
