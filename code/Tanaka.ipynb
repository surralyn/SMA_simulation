```python
import numpy as np
import matplotlib.pyplot as plt
from Tanaka import SMA

my_SMA = SMA('../materials/sma_property.yml')
```

```python
plt.figure(dpi=150)
Ts = [23, 40, 60]
strains = [0.08, 0.06, 0.04, 0.02]
for strain in strains:
    m = my_SMA.curve(60, strain)
    plt.plot(m['strain'], m['stress'], lw=0.8)
plt.legend(strains, loc='lower right')
plt.xlabel('strain')
plt.ylabel('stress/MPa')
plt.axhline(0, c='red', lw=0.5)
plt.axvline(0, c='red', lw=0.5)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(dpi=180)
ax = Axes3D(fig)

T = 60
m = my_SMA.curve(T, 0.08)
stress = m['stress']
strain = m['strain']
Ts = np.ones(len(stress)) * T
ax.plot(strain, Ts, stress)

T = 50
m = my_SMA.curve(T, 0.075)
stress = m['stress']
strain = m['strain']
Ts = np.ones(len(stress)) * T
ax.plot(strain, Ts, stress)

T = 40
m = my_SMA.curve(T, 0.075)
stress = m['stress']
strain = m['strain']
Ts = np.ones(len(stress)) * T
ax.plot(strain, Ts, stress)

T = 30
m = my_SMA.curve(T, 0.07)
stress = m['stress']
strain = m['strain']
Ts = np.ones(len(stress)) * T
ax.plot(strain, Ts, stress)

T = 20
m = my_SMA.curve(T, 0.07)
stress = m['stress']
strain = m['strain']
Ts = np.ones(len(stress)) * T
ax.plot(strain, Ts, stress)

ax.set_xlabel('strain')
ax.set_ylabel('T/')
ax.set_zlabel('stress/MPa')
plt.show()
```
